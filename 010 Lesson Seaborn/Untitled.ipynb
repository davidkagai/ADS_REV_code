# Styling and Themes in Seaborn

```python
# Let us try applying these themes
import matplotlib.pyplot as plt 

#load dataset tips 
tips = sns.load_dataset('tips')

# set theme 
sns.set_style('darkgrid')


#plot 
sns.stripplot(x=tips["day"],y='total_bill',data=tips )
plt.xlabel("Days of the week")
plt.ylabel("Total Payment")

tips.head()
```

# Seaborn’s plotting functions

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

# ECDF

```python
penguins = sns.load_dataset("penguins")

penguins
```

```python
#ECDF represents the proportion or count of observations falling below each unique value in a dataset. 
sns.ecdfplot(data=penguins,x='body_mass_g')
```

# Categorical plots

```python
exercise = sns.load_dataset("exercise")

exercise.head()
```

```python
sns.catplot(x='time',y='pulse',data=exercise,kind='box')
```

```python
sns.boxplot(x='time',y='pulse',data=exercise)
```
